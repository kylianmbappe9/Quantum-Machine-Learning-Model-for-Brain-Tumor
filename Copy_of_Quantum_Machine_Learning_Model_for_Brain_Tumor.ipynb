defining the quanutm model

```python
import pennylane as qml
from pennylane import numpy as np

def quantum_classifer(images, weights):
    qml.templates.AngleEmbedding(images, wires=range(4))
    qml.templates.BasicEntanglerLayers(weights, wires=range(4))
    return qml.expval(qml.PauliZ(0)) and qml.expval(qml.PauliZ(1))
```
